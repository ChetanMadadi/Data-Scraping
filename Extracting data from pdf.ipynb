```python
import camelot as cam
import os                        # provides functions for creating and removing a directory (folder), fetching its contents, changing and identifying the current directory
import pandas as pd              # flexible open source data analysis/manipulation tool
import glob                      # generates lists of files matching given patterns
import pdfplumber                # extracts information from .pdf documents

"""
Obtain key words from repetitive documents, then extract as a dataframe to an .xlsx !
"""

# defining the functions used in main()
def get_keyword(start, end, text):
    """
    start: should be the word prior to the keyword.
    end: should be the word that comes after the keyword.
    text: represents the text from the page(s) you've just extracted.
    """
    for i in range(len(start)):
        try:
            field = ((text.split(start[i]))[1].split(end[i])[0])
            return field
        except:
            continue
```

```python
import numpy as np
folder_to_work = "Temp"
df = pd.DataFrame()
df1 = pd.DataFrame()
df2 = pd.DataFrame()
df3 = pd.DataFrame()
df4 = pd.DataFrame()
for file in os.listdir(folder_to_work):
    file = "/content/Temp/" + file
    start = ['Institute Name: ']
    end = [' [I']
    with pdfplumber.open(file) as pdf:
            page = pdf.pages[0]
            text = page.extract_text()
            text = " ".join(text.split())
    keyword = get_keyword(start, end, text)
    table = cam.read_pdf(file, pages = "all")
    # Total Actual Student Strength
    list = []
    program_table = table[1].df.drop(0)
    for i in range (len(program_table.index)):
      list.append(keyword)
    list
    program_table.insert(0, "", list, True)
    df = df.append(program_table)
    # Financial Resources: Utilised Amount for the Capital expenditure
    list = []
    program_table = table[-5].df
    if len(program_table.index) > 3:
      program_table.drop(program_table.index[[0,1,2]], inplace=True)
    for i in range (len(program_table.index)):
      list.append(keyword)
    list
    program_table.insert(0, "", list, True)
    df1 = df1.append(program_table)
    # Financial Resources: Utilised Amount for the Operational expenditure
    list = []
    program_table = table[-4].df
    if len(program_table.index) > 3:
      program_table.drop(program_table.index[[0,1,2]], inplace=True)
    for i in range (len(program_table.index)):
      list.append(keyword)
    list
    program_table.insert(0, "", list, True)
    df2 = df2.append(program_table)
    # Sponsored Research Details
    list = []
    program_table = table[-3].df.drop(0)
    for i in range (len(program_table.index)):
      list.append(keyword)
    list
    program_table.insert(0, "", list, True)
    df3 = df3.append(program_table)
    # Consultancy Project Details
    list = []
    program_table = table[-2].df.drop(0)
    for i in range (len(program_table.index)):
      list.append(keyword)
    list
    program_table.insert(0, "", list, True)
    df4 = df4.append(program_table)
df2
```

```python
with pd.ExcelWriter('NIRF_2022_Law.xlsx') as writer:  
     df.to_excel(writer, sheet_name='Total Actual Student Strength')
     df1.to_excel(writer, sheet_name='Financial Resources Capital')
     df2.to_excel(writer, sheet_name='Financial Resources Operational')
     df3.to_excel(writer, sheet_name='Sponsored Research Details')
     df4.to_excel(writer, sheet_name='Consultancy Project Details')
```
